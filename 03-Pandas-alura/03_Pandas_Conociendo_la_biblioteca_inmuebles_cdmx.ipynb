````python
# Creando una funcion "Robusta" , de manera que pasando parametros podamos
# hacerlo (Gemini)
from typing import List, Union

def agregar_datos(
    df: pd.DataFrame,
    columnas_agrupar: Union[str, List[str]],
    columna_valor: str,
    operacion: str = 'mean'  # Agregamos flexibilidad
) -> pd.DataFrame:
    """
    Ejecuta una agregación agrupada.
    """
    # Aplicamos el split-apply-combine / Evitamos el Sort si no es necesario
    resultado = df.groupby(by=columnas_agrupar, sort=False)[[columna_valor]].agg(operacion)
    return resultado

# 1. Llamada a la función
df_tipo_precio_2 = agregar_datos(datos, 'Tipo', 'Valor', operacion='mean')

# 2. Ordenamiento (Pipeline de procesamiento)
# Recordando que sort=False en groupby es más eficiente,
# ordenamos al final solo el set resultante (mucho más pequeño).
df_tipo_precio_2 = df_tipo_precio_2.sort_values('Valor')

df_tipo_precio_2.plot(kind='barh', figsize=(10,7),color='purple')

````

##### Parámetros Principales del Método `groupby`

Este método cuenta con varios parámetros que permiten ajustar la eficiencia y el comportamiento de la agrupación:

* **`by`**: Es el parámetro más común. Se utiliza para especificar la columna o columnas por las cuales queremos agrupar. Como argumento, podemos pasar el nombre de una **columna (string)** o una **lista de nombres (list)** para agrupaciones multinivel.
* **`axis`**: Especifica el eje a lo largo del cual se realiza la agrupación.
    * `0` (predeterminado): Agrupa por **filas**.
    * `1`: Agrupa por **columnas**.
* **`sort`**: Parámetro booleano (`True` o `False`). Indica si los grupos resultantes deben ordenarse por el valor de la columna de agrupación.
    * *Nota de rendimiento:* Establecerlo en `False` puede mejorar la velocidad en datasets masivos.
* **`dropna`**: Controla si los valores faltantes (**NaN**) se eliminan o se mantienen como un grupo durante el proceso. El valor predeterminado es `True` (los elimina).

#### 🧠 Análisis de Arquitectura: Pipeline de Limpieza Optimizado

Este bloque de código representa una refactorización basada en principios de **Eficiencia Computacional** y **Clean Code**. A diferencia del enfoque imperativo (paso a paso con variables intermedias), aquí utilizamos **Method Chaining** (encadenamiento) para crear un flujo de transformación de datos lógico y eficiente.

**Ventajas Técnicas del Orden de Ejecución:**

1.  **Filtrado Temprano (`.query`):** Aplicamos la reducción de filas al inicio. Si el dataset original tenía 1M de filas y "Departamento" son solo 20k, liberamos inmediatamente la memoria de 980k registros innecesarios antes de procesar nada más.
2.  **Restricciones Duras (`.dropna`):** Eliminamos los datos estructuralmente inválidos *antes* de intentar arreglarlos. Evita gastar ciclos de CPU imputando ceros en filas que serán descartadas por falta de información crítica (Valor/Condominio).
3.  **Reducción Vertical (`.drop`):** Eliminamos la columna `Tipo` *antes* de la imputación final. Esto evita que la función `.fillna()` tenga que escanear y procesar una columna de texto que ya no aporta valor, ahorrando overhead de procesamiento.
4.  **Imputación Residual (`.fillna`):** Solo rellenamos los huecos en los datos que han sobrevivido a todos los filtros de calidad anteriores.

````python
# Esta es la version optimizada
df_departamentos_optimizado = (
    pd.read_csv(url, sep=';')           # 1. Ingesta (Carga en RAM)
    .query("Tipo == 'Departamento'") # 2. Filtrado (Reduce filas drásticamente)
    .dropna(subset=['Valor', 'Condominio']) # 3. Poda (Elimina datos estructuralmente inválidos)
    .drop(columns=['Tipo'])     # 4. Reducción Vertical (Libera memoria de columnas no usadas)
    .fillna(0)                  # 5. Imputación (Rellena huecos en filas válidas restantes)
)
df_departamentos_optimizado.head()
````

### 🛠️ Métodos para Crear Columnas en Pandas

Existen diversas estrategias para enriquecer un DataFrame dependiendo del contexto (rendimiento vs. legibilidad). A continuación, se presentan las 4 formas más comunes:

**1. Asignación directa de valores**

Inyección de una lista o array de longitud igual al índice del DataFrame.
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
```

**2. Operaciones Vectorizadas (Aritmética de Columnas)**

Calcula valores basándose en operaciones matemáticas entre columnas existentes. Es el método más eficiente en rendimiento (uso de CPU).
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
```

**3. Uso del método assign()**

Permite crear nuevas columnas dentro de un flujo de encadenamiento de métodos (Method Chaining). Retorna un nuevo DataFrame
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
```
**4. Uso del método apply() (Funcional)**

Aplica una función arbitraria a cada elemento (o fila) de una columna. Es útil para lógica compleja que no se puede vectorizar fácilmente.
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)
```
