````python
# 1. Nuestra lista de datos inicial: tuplas (nombre, calificacion)
resultados_estudiantes = [
    ("Ana", 7),
    ("Juan", 9),
    ("María", 6),
    ("Pedro", 8),
    ("Sofía", 10)
]

print(f"Lista de resultados inicial: {resultados_estudiantes}\n")

# 2. Usamos map y lambda para extraer la calificación y elevarla al cuadrado
#    La función lambda `item: item[1]**2` toma cada tupla (item)
#    y devuelve el segundo elemento (item[1], que es la calificación) elevado al cuadrado.
calificaciones_cuadrado_map = map(lambda item: item[1]**2, resultados_estudiantes)

print(f"Objeto map con calificaciones al cuadrado: {calificaciones_cuadrado_map}")

# 3. Convertimos el objeto map a una lista para ver los resultados y poder usarlos fácilmente
calificaciones_cuadrado_lista = list(calificaciones_cuadrado_map)

print(f"Lista de calificaciones al cuadrado: {calificaciones_cuadrado_lista}\n")


# 4. Ahora calculamos el promedio de estas calificaciones al cuadrado.
#    Para el promedio, necesitamos la suma de los elementos y el conteo de los elementos.
#    Mientras que `map` se usa para transformar elementos individualmente,
#    funciones como `sum()` y `len()` se usan para operaciones de agregación.
#    No es común usar una 'lambda' con 'map' para calcular un promedio directamente
#    porque 'map' opera elemento por elemento, no agrega.

suma_cuadrados = sum(calificaciones_cuadrado_lista)
numero_estudiantes = len(calificaciones_cuadrado_lista)

# Verificamos que no haya división por cero por si la lista estuviera vacía
if numero_estudiantes > 0:
    promedio_cuadrados = suma_cuadrados / numero_estudiantes
    print(f"Suma de las calificaciones al cuadrado: {suma_cuadrados}")
    print(f"Número de estudiantes: {numero_estudiantes}")
    print(f"Promedio de las calificaciones al cuadrado: {promedio_cuadrados:.2f}")
else:
    print("No hay estudiantes para calcular el promedio.")
````

### Resumen de este ejemplo:

*   **`resultados_estudiantes`**: Nuestra lista de tuplas original.
*   **`map(lambda item: item[1]**2, resultados_estudiantes)`**: Aquí es donde reside la magia de `map` y `lambda`:
    *   `lambda item: item[1]**2`: Esta función `lambda` recibe cada `item` (una tupla como `("Ana", 7)`). Dentro de la lambda, `item[1]` accede a la calificación (el segundo elemento de la tupla) y luego `**2` la eleva al cuadrado.
    *   `map(...)`: Aplica esta `lambda` a cada tupla en `resultados_estudiantes`, generando un objeto `map` que contiene los cuadrados de las calificaciones.
*   **`list(calificaciones_cuadrado_map)`**: Como antes, convertimos el objeto `map` a una lista (`calificaciones_cuadrado_lista`) para poder manipular los resultados finales.
*   **`sum(...)` y `len(...)`**: Una vez que tenemos la lista de calificaciones al cuadrado, usamos las funciones incorporadas `sum()` y `len()` para calcular el promedio. Es importante destacar que `map` está diseñado para *transformar* elementos individualmente, no para *agregar* o resumir una colección completa de datos como lo haría un promedio. Para las agregaciones, se usan funciones como `sum()`, `len()`, `min()`, `max()`, o librerías como `pandas` para conjuntos de datos más grandes.

### 3. ¿Qué sucede si hay un elemento intermedio en la lista de nombres o de apellidos que está vacío?

Si hay una cadena vacía o una cadena con solo espacios en blanco en la lista original, el comportamiento de `.title()` o `.capitalize()` y la posterior concatenación sería el siguiente:

*   **`''.title()`** o **`''.capitalize()`**: Ambas funciones devuelven una cadena vacía `''`.
*   **`'   '.title()`** o **`'   '.capitalize()`**: Ambas funciones devuelven una cadena vacía `''` después de aplicar la capitalización, ya que no hay caracteres alfanuméricos para capitalizar.

Cuando concatenas con un f-string, un nombre o apellido vacío resultará en un espacio extra o en un elemento que solo contiene el otro componente (nombre o apellido). El `.strip()` al final de la lambda en el ejemplo anterior (`.strip()`) es crucial para eliminar cualquier espacio en blanco adicional si uno de los componentes está vacío.

Veamos un ejemplo:

````python
nombres_con_vacios = ["ana", "", "carlos"]
apellidos_con_vacios = ["gomez", "perez", ""]

# Normalizar
nombres_normalizados_vacios = list(map(lambda x: x.title(), nombres_con_vacios))
apellidos_normalizados_vacios = list(map(lambda x: x.title(), apellidos_con_vacios))

# Concatenar usando zip (se detendrá con la lista más corta)
nombres_completos_vacios_zip = list(map(
    lambda nombre_apellido: f"{nombre_apellido[0]} {nombre_apellido[1]}".strip(),
    zip(nombres_normalizados_vacios, apellidos_normalizados_vacios)))

print(f"Nombres originales (con vacíos): {nombres_con_vacios}")
print(f"Apellidos originales (con vacíos): {apellidos_con_vacios}")
print(f"Nombres completos (con zip, usando .strip()): {nombres_completos_vacios_zip}\n")


# Concatenar usando zip_longest para ver todos los elementos, incluyendo los None/empty
nombres_completos_vacios_zip_longest = list(map(
    lambda nombre_apellido: f"{nombre_apellido[0] or ''} {nombre_apellido[1] or ''}".strip(),
    zip_longest(nombres_normalizados_vacios, apellidos_normalizados_vacios, fillvalue=''))
)

print(f"Nombres completos (con zip_longest, usando .strip()): {nombres_completos_vacios_zip_longest}")
````

8 - Como científico de datos en un equipo de fútbol, necesitas implementar nuevas formas de recopilación de datos sobre el rendimiento de los jugadores y del equipo en su conjunto. Tu primera acción es crear una forma de calcular la puntuación del equipo en el campeonato nacional a partir de los datos de goles marcados y recibidos en cada juego.

Escribe una función llamada calcula_puntos() que recibe como parámetros dos listas de números enteros, representando los goles marcados y recibidos por el equipo en cada partido del campeonato. La función debe devolver la puntuación del equipo y el rendimiento en porcentaje, teniendo en cuenta que la victoria vale 3 puntos, el empate 1 punto y la derrota 0 puntos.

Nota: si la cantidad de goles marcados en un partido es mayor que los recibidos, el equipo ganó. En caso de ser igual, el equipo empató, y si es menor, el equipo perdió. Para calcular el rendimiento, debemos hacer la razón entre la puntuación del equipo y la puntuación máxima que podría recibir.

Para la prueba, utiliza las siguientes listas de goles marcados y recibidos:

````python
goles_marcados = [2, 1, 3, 1, 0]
goles_recibidos = [1, 2, 2, 1, 3]
# Texto probablemente mostrado:
# La puntuación del equipo fue `puntos` y su rendimiento fue `desempeno`%"
````

````python
def calcula_puntos(goles_marcados, goles_recibidos):
    puntuacion_total = 0
    partidos_jugados = len(goles_marcados)
    puntuacion_maxima_posible = partidos_jugados * 3  # Cada partido es una posible victoria (3 puntos)

    for i in range(partidos_jugados):
        goles_a_favor = goles_marcados[i]
        goles_en_contra = goles_recibidos[i]

        if goles_a_favor > goles_en_contra:
            puntuacion_total += 3  # Victoria
        elif goles_a_favor == goles_en_contra:
            puntuacion_total += 1  # Empate
        else:
            puntuacion_total += 0  # Derrota

    # Calcular el rendimiento en porcentaje
    if puntuacion_maxima_posible > 0:
        rendimiento = (puntuacion_total / puntuacion_maxima_posible) * 100
    else:
        rendimiento = 0.0 # No se jugaron partidos

    return puntuacion_total, rendimiento

# Para la prueba, utiliza las siguientes listas de goles marcados y recibidos:
goles_marcados = [2, 1, 3, 1, 0]
goles_recibidos = [1, 2, 2, 1, 3]

# Usar la función con los datos de prueba
puntos, desempeno = calcula_puntos(goles_marcados, goles_recibidos)

# Mostrar el mensaje final
print(f"La puntuación del equipo fue {puntos} y su rendimiento fue {desempeno:.2f}%")
````

9 - Te han desafiado a crear un código que calcule los gastos de un viaje a una de las cuatro ciudades desde Recife, siendo ellas: Salvador, Fortaleza, Natal y Aracaju.

El costo diario del hotel es de 150 reales en todas ellas y el consumo de gasolina en el viaje en coche es de 14 km/l, siendo que el precio de la gasolina es de 5 reales por litro. Los gastos con paseos y alimentación a realizar en cada una de ellas por día serían [200, 400, 250, 300], respectivamente.

Sabiendo que las distancias entre Recife y cada una de las ciudades son aproximadamente [850, 800, 300, 550] km, crea tres funciones: la primera función calcula los gastos de hotel (gasto_hotel), la segunda calcula los gastos de gasolina (gasto_gasolina) y la tercera los gastos de paseo y alimentación (gasto_paseo).

Para probar, simula un viaje de 3 días a Salvador desde Recife. Considera el viaje de ida y vuelta en coche.

````python
# Definición de las constantes y datos del problema
COSTO_DIARIO_HOTEL = 150  # Reales
CONSUMO_KM_LITRO = 14    # km/l
PRECIO_LITRO_GASOLINA = 5 # Reales/litro

CIUDADES = ["Salvador", "Fortaleza", "Natal", "Aracaju"]
GASTOS_PASEO_ALIMENTACION_DIARIO = [200, 400, 250, 300] # Reales por día, por ciudad
DISTANCIAS_RECIFE = [850, 800, 300, 550] # km, desde Recife

# --- Funciones para calcular gastos ---

def gasto_hotel(dias):
  """Calcula el costo total del hotel para un número de días."""
  return dias * COSTO_DIARIO_HOTEL

def gasto_gasolina(distancia_ida):
  """Calcula el costo total de gasolina para un viaje de ida y vuelta."""
  distancia_total = distancia_ida * 2  # Viaje de ida y vuelta
  litros_necesarios = distancia_total / CONSUMO_KM_LITRO
  costo_total_gasolina = litros_necesarios * PRECIO_LITRO_GASOLINA
  return costo_total_gasolina

def gasto_paseo(dias, gasto_diario_ciudad):
  """Calcula el costo total de paseos y alimentación para un número de días en una ciudad específica."""
  return dias * gasto_diario_ciudad

# --- Simulación del viaje con entrada de usuario ---

try:
    DIAS_VIAJE = int(input("¿Cuántos días durará el viaje? "))
    CIUDAD_DESTINO = input("¿A qué ciudad desea viajar (Salvador, Fortaleza, Natal, Aracaju)? ")

    # Encontrar el índice de la ciudad de destino en las listas para obtener sus datos
    indice_ciudad = CIUDADES.index(CIUDAD_DESTINO.capitalize()) # .capitalize() para coincidir con la lista CIUDADES
    distancia_a_ciudad = DISTANCIAS_RECIFE[indice_ciudad]
    gasto_diario_ciudad = GASTOS_PASEO_ALIMENTACION_DIARIO[indice_ciudad]

    # Calcular los gastos individuales
    costo_hotel = gasto_hotel(DIAS_VIAJE)
    costo_gas = gasto_gasolina(distancia_a_ciudad)
    costo_paseos = gasto_paseo(DIAS_VIAJE, gasto_diario_ciudad)

    # Calcular el gasto total
    gastos_totales = costo_hotel + costo_gas + costo_paseos

    # Mostrar el resultado
    print(f"\nCon base en los gastos definidos, un viaje de {DIAS_VIAJE} días a {CIUDAD_DESTINO} desde Recife costaría {gastos_totales:.2f} reales.")

except ValueError:
    print(f"Error: Asegúrate de introducir un número válido para los días y una ciudad válida (Salvador, Fortaleza, Natal o Aracaju).")
except IndexError:
    print(f"Error: La ciudad '{CIUDAD_DESTINO}' no se encontró en la lista de ciudades válidas.")
````

````python
#Solucion Instructor ALura Ejericio 9
dias = int(input("¿Cuántas diarias? "))
ciudad = input("¿Cuál es la ciudad? [Salvador, Fortaleza, Natal o Aracaju]: ")
distancias = [850, 800, 300, 550]
paseo = [200, 400, 250, 300]
km_l = 14
gasolina = 5

def gasto_hotel(dias):
    return 150 * dias

def gasto_gasolina(ciudad):
    if ciudad == "Salvador":
        return (2 * distancias[0] * gasolina) / km_l
    elif ciudad == "Fortaleza":
        return (2 * distancias[1] * gasolina) / km_l
    elif ciudad == "Natal":
        return (2 * distancias[2] * gasolina) / km_l
    elif ciudad == "Aracaju":
        return (2 * distancias[3] * gasolina) / km_l

def gasto_paseo(ciudad, dias):
    if ciudad=="Salvador":
        return paseo[0] * dias
    elif ciudad=="Fortaleza":
        return paseo[1] * dias
    elif ciudad=="Natal":
        return paseo[2] * dias
    elif ciudad=="Aracaju":
        return paseo[3] * dias

gastos = gasto_hotel(dias) + gasto_gasolina(ciudad) + gasto_paseo(ciudad, dias)
print(f"Con base en los gastos definidos, un viaje de {dias} días a {ciudad} desde Recife costaría {round(gastos, 2)} reales")
````

10 - Has comenzado una pasantía en una empresa que trabaja con procesamiento de lenguaje natural (NLP). Tu líder te solicitó que crees un fragmento de código que reciba una frase escrita por el usuario y filtre solo las palabras con un tamaño mayor o igual a 5, mostrándolas en una lista. Esta demanda se centra en el análisis del patrón de comportamiento de las personas al escribir palabras de esta cantidad de caracteres o más.

Consejo: utiliza las funciones lambda y filter() para filtrar estas palabras. Recordando que la función integrada filter() recibe una función (en nuestro caso, una función lambda) y filtra un iterable según la función. Para tratar la frase, utiliza replace() para cambiar ',' '.', '!' y '?' por espacio.

Utiliza la frase "Aprender Python aquí en Alura es muy bueno" para probar el código.

````python
frase_original = "Aprender Python aquí en Alura es muy bueno"

# Puntuaciones a reemplazar por espacios
puntuaciones = [',', '.', '!', '?']

# Limpiar la frase de puntuaciones
frase_limpia = frase_original
for punc in puntuaciones:
    frase_limpia = frase_limpia.replace(punc, ' ')

# Dividir la frase limpia en palabras
palabras = frase_limpia.split()

# Filtrar las palabras con longitud mayor o igual a 5 usando filter() y lambda
palabras_filtradas_map = filter(lambda palabra: len(palabra) >= 5, palabras)

# Convertir el objeto filter a una lista para mostrar el resultado
palabras_resultantes = list(palabras_filtradas_map)

print(f"Frase original: '{frase_original}'")
print(f"Frase limpia (sin puntuaciones): '{frase_limpia}'")
print(f"Palabras filtradas (>= 5 caracteres): {palabras_resultantes}")
````

````python
#Resolucio  instructor Alura:
# Solicitando una frase y separándola por espacios. Usando replace para cambiar
# puntuaciones por espacios.
frase = input("Escribe una frase: ")
frase = frase.replace(',',' ').replace('.',' ').replace('!',' ').replace('?',' ').split()

# Filtrando la frase en formato de lista, pasando a la lista tamaño
# solo las palabras con 5 o más caracteres e imprimiéndola en pantalla
tamano = list(filter(lambda x: len(x) >= 5, frase))
print(tamano)
````
