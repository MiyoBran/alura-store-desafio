```python
#1 - Escribe un programa que pida a la persona usuaria que proporcione dos números y muestre el número más grande.
print("Ingrese dos numeros , para determinar cual es el mayor: ")
numero1 = int(input("Ingrese el primer numero: "))
numero2 = int(input("Ingrese el segundo numero: "))
if numero1 > numero2:
  print(f"El numero mayor es: {numero1}")
else:
  print(f"El numero mayor es: {numero2}")


# 2 - Escribe un programa que solicite el porcentaje de crecimiento de producción de una empresa e informe si hubo un crecimiento (porcentaje positivo) o una disminución (porcentaje negativo).
print('Ejercicio crecimiento de produccion: ')
produccion_anterior = float(input("Ingrese la produccion del periodo anterior: "))
produccion_actual = float(input("Ingrese la produccion del periodo actual: "))
porcentaje_crecimiento = (produccion_actual - produccion_anterior) / produccion_anterior * 100
if porcentaje_crecimiento > 0:
  print(f"Hubo un crecimiento del {porcentaje_crecimiento}%")
else:
  print(f"Hubo una disminucion del {porcentaje_crecimiento}%")
```
