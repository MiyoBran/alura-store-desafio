````python
def agrupar_listas(lista_1, lista_2):
    # 1. Validación de longitud (Regla de negocio)
    if len(lista_1) != len(lista_2):
        raise IndexError("La cantidad de elementos en cada lista es diferente.")

    lista_tuplas = []
    try:
        # 2. Bucle usando zip
        for a, b in zip(lista_1, lista_2):
            lista_tuplas.append((a, b, a + b))

        # Solo llegamos aquí si el bucle terminó sin errores
        return lista_tuplas

    except TypeError as e:
        print(f"--- ALERTA DE SISTEMA ---")
        print(f"Error crítico de tipos: No se puede sumar un texto con un número.")
        print(f"Detalle técnico: {e}")

        # Opción 2: Re-lanzamos el error para que el programa principal se detenga
        raise

    finally:
        # Esto se ejecutará SIEMPRE, ayudando a la auditoría del sistema
        print("Log: Intento de ejecución de agrupar_listas finalizado.")

# Prueba con error
lista_A = [4, 6, 7, 9, 10]
lista_error = [4, 7, 9.5, 10, 'A']

try:
    resultado = agrupar_listas(lista_A, lista_error)
    print(resultado)
except TypeError:
    print("\n[Resultado Final]: La operación fue abortada para proteger la integridad de los datos.")
````

**5 -** Como desafío, se te ha asignado la tarea de desarrollar un código que contabiliza las puntuaciones de estudiantes de una institución educativa de acuerdo con sus respuestas en una prueba. Este código debe ser probado para un ejemplo de 3 estudiantes con una lista de listas en la que cada lista tiene las respuestas de 5 preguntas objetivas de cada estudiante. Cada pregunta vale un punto y las alternativas posibles son A, B, C o D.

Si alguna alternativa en una de las pruebas no está entre las alternativas posibles, debes lanzar un `ValueError` con el mensaje "La alternativa [alternativa] no es una opción de alternativa válida". El cálculo de las 3 notas solo se realizará mediante las entradas con las alternativas A, B, C o D en todas las pruebas. Si no se lanza la excepción, se mostrará una lista con las notas en cada prueba.

**Datos para la prueba del código:**

*Respuestas de la prueba:*

```python
respuestas = ['D', 'A', 'B', 'C', 'A']

```

*A continuación, hay 2 listas de listas que puedes usar como prueba:*

*Notas sin excepción:*

```python
tests_sin_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'C', 'A'], ['D', 'B', 'A', 'C', 'A']]

```

*Notas con excepción:*

```python
tests_con_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'E', 'A'], ['D', 'B', 'A', 'C', 'A']]

```

````python

def evaluar_respuestas(pruebas: list):
    respuestas_correctas = ['D', 'A', 'B', 'C', 'A']
    puntuaciones = []

    try:
        for prueba in pruebas:
            nota = 0  # Reiniciamos la nota para cada estudiante

            for i in range(len(prueba)):
                # 1. Validación de opciones permitidas
                if prueba[i] not in ['A', 'B', 'C', 'D']:
                    raise ValueError(f"La alternativa {prueba[i]} no es una opción de alternativa válida.")

                # 2. Conteo de puntos
                elif prueba[i] == respuestas_correctas[i]:
                    nota += 1

            # 3. Guardamos la nota final del estudiante tras evaluar sus 5 respuestas
            puntuaciones.append(nota)

    except ValueError as e:
        print(f"Error detectado: {e}")
        return None
    except IndexError:
        print("Error: El examen no tiene la cantidad de preguntas requerida.")
        return None
    else:
        # Se ejecuta solo si no hubo ninguna excepción
        return puntuaciones
````

````python
# Datos para la prueba del código
respuestas_correctas = ['D', 'A', 'B', 'C', 'A']

# Caso 1: Notas sin excepción (Todas las opciones son A, B, C o D)
tests_sin_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'C', 'A'], ['D', 'B', 'A', 'C', 'A']]

# Caso 2: Notas con excepción (Contiene una 'E' en el segundo examen)
tests_con_ex = [['D', 'A', 'B', 'C', 'A'], ['C', 'A', 'A', 'E', 'A'], ['D', 'B', 'A', 'C', 'A']]

print("Prueba Sin Excepción:", evaluar_respuestas(tests_sin_ex))

print("\n")
print("\nPrueba Con Excepción:", evaluar_respuestas(tests_con_ex))
````

````python
#Refactoring Usando zip
def evaluar_con_zip(pruebas):
    patron = ['D', 'A', 'B', 'C', 'A']
    reporte_final = []

    try:
        for examen in pruebas:
            if len(examen) != len(patron):
                raise IndexError("Longitud de examen incorrecta.")

            # Usamos zip para comparar respuesta por respuesta
            # zip(examen, patron) crea parejas: (Respuesta_Alumno, Respuesta_Correcta)
            nota = sum(1 for r_alumno, r_correcta in zip(examen, patron)
                       if (r_alumno in 'ABCD' or (raise_val_err := r_alumno))
                       and r_alumno == r_correcta)

            # Nota: Para mantener tu validación de ValueError dentro de un sum/zip
            # se suele usar un bucle más explícito para mayor claridad:
            nota_limpia = 0
            for r_alumno, r_correcta in zip(examen, patron):
                if r_alumno not in ['A', 'B', 'C', 'D']:
                    raise ValueError(f"Alternativa inválida detectada: {r_alumno}")
                if r_alumno == r_correcta:
                    nota_limpia += 1

            reporte_final.append(nota_limpia)

    except (ValueError, IndexError) as e:
        print(f"--- FALLO DE VALIDACIÓN ZIP ---")
        print(f"Causa: {e}")
        return None
    else:
        return reporte_final
    finally:
        print("Log: Auditoría por ZIP concluida.")

# Ejecución de prueba con ZIP
print("Prueba ZIP Sin Excepción:", evaluar_con_zip(tests_sin_ex))
print("\nPrueba ZIP Con Excepción:", evaluar_con_zip(tests_con_ex))
````

````python
def evaluar_eliminacion_puntuacion(lista_palabras: list):
    puntuacion = [',', '.', '!', '?']

    try:
        for palabra in lista_palabras:
            # Opción recomendada por legibilidad
            if any(caracter in palabra for caracter in puntuacion):
                raise ValueError(f'El texto presenta puntuaciones en la palabra "{palabra}"')

        # Si termina el bucle sin errores
        return "Lista tratada correctamente."

    except ValueError as e:
        # Aquí capturamos el raise y lo mostramos
        return f"Error de validación: {e}"

# --- Bloque de Pruebas ---
lista_no_tratada = ['Python', 'es', 'un', 'lenguaje', 'de', 'programación', 'poderoso,', 'versátil']

print(evaluar_eliminacion_puntuacion(lista_no_tratada))
````
