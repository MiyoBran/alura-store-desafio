# Pandas


## 1. **Importar biblioteca y base de datos**

### - Conociendo los datos

```python
# Importamos la biblioteca pandas , y seleccionamos el url origen de los datos
import pandas as pd
url = 'https://gist.githubusercontent.com/ahcamachod/a572cfcc2527046db93101f88011b26e/raw/ffb13f45a79d31223e645611a119397dd127ee3c/alquiler.csv'
pd.read_csv(url)
```

```python
# Como el separador no es una coma (por defecto del csv) , debemos indicar que
# el separador es un ';'
pd.read_csv(url, sep=';')
```

```python
#ahora almacenaremos lo leido en una variable
datos = pd.read_csv(url, sep=';')
print(datos)
```

```python
#tenemos mejores formas de visualizar los datos usando head o tail
datos.head()
```

```python
datos.tail()
```

```python
# analogamente podemos determinar la cantidad de filas a visualizar con el
# metodo head o tail , poniendo la cantidad de filas a visualizar como parametro
# de la funcion
datos.head(10)
```

```python
# Si queremos un ejemplo de lo que hay , con sample() tomaremos una muestra aleatoria
datos.sample(10)
```

```python
# Aqui veremos el tipo de variable que es datos
type(datos)
```

### - Explorar las caracteristicas generales de los datos

```python
# Siempre las primeras aproximaciones para un dataframe son tener un ejemplo de
# los datos , y sobre todo el tamaño, para esto usaremos sample() y shape().
datos.sample(5)
```

```python
# Con esto vemos la cantidad de datos seguido de la cantidad de columnas
datos.shape
```

```python
# Si quiero saber cuales son las columnas (Devuelve un objeto de tipo pandas
# INDEX)
datos.columns
```

```python
# SI quiero conocer de forma resumida que hay en mi conjunto de datos , tenemos
# a disposicion el metodo .info
datos.info()
# esto nos dira que es un objeto de tipo DataFrame de pandas , con la cantidad
# de entradas , columnas , ademas con los datatypes , y cuantos elementos no
# nulos
```

```python
#Si queremos ver que tiene una columa
datos['Tipo']
# en este caso nos devuelve una serie pandas, y nos indica debajo datos basicos
# de la misma.
```

```python
# Y en caso de que querramos visualizar mas de una columna , debemos usar dos
# corchetes (le paso una lista con mis dos columnas)
datos[['Habitaciones', 'Valor']]
#En este caso no devuelve una serie pandas , sino un dataframe
```
